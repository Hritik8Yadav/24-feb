```python
#Ans10:
data_fr=pd.DataFrame({'Sales':[22,132,241,532,134,523,122,4234,1322,5212,3244],'Date':['1-1-2023','1-2-2023','1-3-2023','1-4-2023','1-5-2023','1-6-2023','1-7-2023','1-8-2023','1-9-2023','1-10-2023','1-11-2023']})
def add_avg():
    MovingAverage=[]
    for i in range(len(data_fr)):
        window=7
        if i<window-1:
            # print(i)
            # MovingAverage.append(None)
            temp1=data_fr['Sales'][0:i+1].mean()
            MovingAverage.append(temp1)
        else:
            st=i-(window-1)
            end=i+1
            temp=data_fr['Sales'][st:end].mean()
            MovingAverage.append(temp)
    data_fr['MovingAverage']=MovingAverage
add_avg()
data_fr
```

```python
#Ans11:
def add_weekday():
    # data_fr['Weekday']=''
    for i in range(len(data_fr)):
        temp_dta=data_fr['Date'].str.split('-')[i][1]
        if int(temp_dta)%7==1:
            data_fr.loc[i,'Weekday']='Sunday'
        elif int(temp_dta)%7==2:
            data_fr.loc[i,'Weekday']='Monday'
        elif int(temp_dta)%7==3:
            data_fr.loc[i,'Weekday']='Tuesday'
        elif int(temp_dta)%7==4:
            data_fr.loc[i,'Weekday']='Wednesday'
        elif int(temp_dta)%7==5:
            data_fr.loc[i,'Weekday']='Thursday'
        elif int(temp_dta)%7==6:
            data_fr.loc[i,'Weekday']='Friday'
        elif int(temp_dta)%7==0:
            data_fr.loc[i,'Weekday']='Saturday'
add_weekday()
data_fr
```

```python
#Ans12:
def select_january_data():
    data_fr['Date'] = pd.to_datetime(data_fr['Date'])
    
    start_date = pd.Timestamp('2023-01-01')
    end_date = pd.Timestamp('2023-01-31')
    
    mask = (data_fr['Date'] >= start_date) & (data_fr['Date'] <= end_date)
    selected_dat = data_fr.loc[mask]
    
    return selected_dat
select_january_data()
```
